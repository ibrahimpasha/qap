## Strategy Tips for the OS/Linux MCQ Section (25 minutes)

**Time Allocation:** With 10-12 questions in 25 minutes you have roughly 2 minutes per question. Don't spend more than 2.5 minutes on any single question.

**Elimination Technique:** Read all four options first. Eliminate answers that are clearly wrong — even removing one option raises your odds from 25% to 33%.

**Skip and Return:** If a question involves a command you don't recognize, skip it and come back. Answer the ones you're confident about first to lock in points.

**Command Questions:** Visualize the command running in a terminal. Think about what flags do and what the expected output format looks like.

**Conceptual Questions:** For OS theory (scheduling, memory), relate the concept to a real-world analogy. For example, Round Robin scheduling is like taking turns in a game.

**Common Traps:** Watch for subtle differences between similar commands (e.g., `netstat` vs `ss`, `chmod 755` vs `chmod 644`). Pay attention to whether the question asks about the command's default behavior or a specific flag.

## Linux File System Hierarchy (Beginner)

### Linux File System Hierarchy

Linux organizes files in a single-rooted tree structure starting at `/`. Every file and directory lives under this root. Understanding the standard directory layout (defined by the Filesystem Hierarchy Standard — FHS) is essential for system administration.

| Directory | Purpose |
|-----------|--------|
| `/` | Root of the entire filesystem |
| `/home` | User home directories (e.g., `/home/alice`) |
| `/etc` | System-wide configuration files |
| `/var` | Variable data — logs, caches, spool files |
| `/tmp` | Temporary files (cleared on reboot on many distros) |
| `/usr` | User programs, libraries, documentation |
| `/bin` | Essential user command binaries (ls, cp, mv) |
| `/sbin` | System administration binaries (fdisk, iptables) |
| `/dev` | Device files (disks, terminals, null device) |
| `/proc` | Virtual filesystem exposing kernel/process info |

**Key points:**
- `/etc` is where you find config files like `/etc/hosts`, `/etc/fstab`, `/etc/passwd`.
- `/var/log` stores system and application logs.
- `/proc` is not on disk — it's generated by the kernel at runtime. For example, `/proc/cpuinfo` shows CPU details.
- `/dev/null` is the "black hole" — anything written to it is discarded.

````python
# --- Exploring the filesystem hierarchy ---
# List top-level directories
import subprocess, textwrap

# Simulated output of: ls /
top_level = textwrap.dedent('''
    $ ls /
    bin   dev  home  lib    media  opt   root  sbin  sys  usr
    boot  etc  init  lib64  mnt    proc  run   srv   tmp  var
''')
print(top_level)

# View CPU info from /proc
proc_cpuinfo = textwrap.dedent('''
    $ cat /proc/cpuinfo | head -5
    processor       : 0
    vendor_id       : GenuineIntel
    cpu family      : 6
    model           : 142
    model name      : Intel(R) Core(TM) i7-8550U CPU @ 1.80GHz
''')
print(proc_cpuinfo)

# Check disk usage of key directories
du_output = textwrap.dedent('''
    $ du -sh /var/log /etc /tmp
    256M    /var/log
    12M     /etc
    4.0K    /tmp
''')
print(du_output)
````

- Know the FHS layout — it's a common interview topic for Linux roles.

## Process Management (Beginner)

### Process Management

A **process** is a running instance of a program. Linux provides powerful tools to monitor, control, and manage processes.

**Key commands:**
- `ps` — snapshot of current processes
- `top` / `htop` — real-time process monitor
- `kill` — send signals to processes
- `nice` / `renice` — set/change process priority
- `bg` / `fg` — move jobs to background/foreground

**Common signals:**
| Signal | Number | Action |
|--------|--------|--------|
| SIGTERM | 15 | Graceful termination (default for `kill`) |
| SIGKILL | 9 | Force kill (cannot be caught or ignored) |
| SIGHUP | 1 | Hangup — often used to reload config |
| SIGSTOP | 19 | Pause process |
| SIGCONT | 18 | Resume paused process |

**Process states:**
- **R** — Running or runnable
- **S** — Sleeping (waiting for an event)
- **D** — Uninterruptible sleep (usually I/O)
- **Z** — Zombie (terminated but not reaped by parent)
- **T** — Stopped (by signal or debugger)

````python
# --- ps: viewing processes ---
import textwrap

ps_output = textwrap.dedent('''
    $ ps aux | head -10
    USER       PID %CPU %MEM    VSZ   RSS TTY  STAT START   TIME COMMAND
    root         1  0.0  0.1 169316 11804 ?    Ss   Jan14   0:05 /sbin/init
    root         2  0.0  0.0      0     0 ?    S    Jan14   0:00 [kthreadd]
    user      1523  0.2  1.5 432100 62340 ?    Sl   10:00   0:15 /usr/bin/python3 app.py
    www-data  1890  0.1  0.3  56780 12340 ?    S    10:05   0:02 nginx: worker process
    user      2001  0.0  0.1  21456  4320 pts/0 Ss  10:10   0:00 bash

    # Show only processes for current user
    $ ps -u $USER
      PID TTY          TIME CMD
     1523 ?        00:00:15 python3
     2001 pts/0    00:00:00 bash
''')
print(ps_output)
````

````python
# --- kill, nice, bg/fg: controlling processes ---
import textwrap

kill_output = textwrap.dedent('''
    # Graceful termination (SIGTERM, default)
    $ kill 1523

    # Force kill (SIGKILL — use as last resort)
    $ kill -9 1523

    # Kill all processes by name
    $ killall python3

    # Start a process with lower priority (higher nice value)
    $ nice -n 10 python3 heavy_task.py &

    # Change priority of running process
    $ renice -n 5 -p 1523
    1523 (process ID) old priority 0, new priority 5

    # Background and foreground
    $ python3 long_task.py
    ^Z                          # Ctrl+Z suspends the process
    [1]+  Stopped  python3 long_task.py
    $ bg                        # resume in background
    [1]+ python3 long_task.py &
    $ fg                        # bring back to foreground
    python3 long_task.py
''')
print(kill_output)
````

What signal does the `kill` command send by default (without specifying a signal)?

**A.** SIGKILL (9)
**B.** SIGTERM (15)
**C.** SIGHUP (1)
**D.** SIGINT (2)

### Solution — Question 1

**Correct Answer: B**

By default, `kill <PID>` sends SIGTERM (signal 15), which requests graceful termination. The process can catch this signal and clean up before exiting.

**Why the other options are incorrect:**

- **A.** SIGKILL (9) must be specified explicitly with `kill -9`. It cannot be caught.
- **C.** SIGHUP (1) is the hangup signal, typically sent when a terminal closes.
- **D.** SIGINT (2) is sent by Ctrl+C, not by the default `kill` command.

### Question 2

A process shows state 'Z' in the `ps` output. What does this mean?

**A.** The process is sleeping, waiting for I/O
**B.** The process is running on the CPU
**C.** The process is a zombie — terminated but not reaped by its parent
**D.** The process is stopped by a debugger

### Solution — Question 2

**Correct Answer: C**

State 'Z' means zombie. The process has finished execution but its parent hasn't called wait() to read its exit status. The process entry remains in the process table until the parent reaps it.

**Why the other options are incorrect:**

- **A.** Sleeping processes show state 'S' (interruptible) or 'D' (uninterruptible).

- `ps aux` shows all running processes; pipe to `grep` to filter.

- `nice` sets priority at launch; `renice` changes priority of a running process.
- Ctrl+Z suspends a process; `bg` resumes it in background; `fg` brings it to foreground.

## File Permissions and Ownership (Beginner)

### File Permissions and Ownership

Linux uses a permission system based on three categories of users and three types of access:

**User categories:**
- **Owner (u)** — the user who owns the file
- **Group (g)** — users in the file's group
- **Others (o)** — everyone else

**Permission types:**
- **r (read = 4)** — view file contents / list directory
- **w (write = 2)** — modify file / add/remove files in directory
- **x (execute = 1)** — run file as program / enter directory

**Numeric notation:**
Each digit is the sum of r(4) + w(2) + x(1):
- `755` = rwxr-xr-x (owner: full, group/others: read+execute)
- `644` = rw-r--r-- (owner: read+write, group/others: read only)
- `700` = rwx------ (owner: full, no access for others)

**umask:** Sets default permissions for new files. Common default umask is `022`, meaning new files get `644` and new directories get `755`.

**Commands:**
- `chmod` — change file permissions
- `chown` — change file owner and/or group

````python
# --- chmod and chown examples ---
import textwrap

perm_output = textwrap.dedent('''
    # View current permissions
    $ ls -l script.sh
    -rw-r--r-- 1 user devteam 2048 Jan 15 10:00 script.sh

    # Make executable (numeric)
    $ chmod 755 script.sh
    $ ls -l script.sh
    -rwxr-xr-x 1 user devteam 2048 Jan 15 10:00 script.sh

    # Make executable (symbolic)
    $ chmod u+x script.sh    # add execute for owner
    $ chmod go-w config.txt  # remove write for group and others
    $ chmod a+r readme.txt   # add read for all

    # Change ownership
    $ chown alice:devteam project/
    $ chown -R alice:devteam project/  # recursive

    # Check umask
    $ umask
    0022
    # New files: 666 - 022 = 644 (rw-r--r--)
    # New dirs:  777 - 022 = 755 (rwxr-xr-x)
''')
print(perm_output)
````

**D.** r--r--r-- (everyone: read only)

### Solution — Question 1

**Correct Answer: B**

6 = r(4)+w(2) = rw-, 4 = r(4) = r--, 4 = r(4) = r--. So 644 = rw-r--r-- — owner can read and write, group and others can only read.

**Why the other options are incorrect:**

- **A.** rwxr--r-- would be 744 (7 = r+w+x), not 644.
- **C.** rw-rw-r-- would be 664, not 644.
- **D.** r--r--r-- would be 444, not 644.

### Question 2

**A.** Removes execute permission for the owner
**B.** Adds execute permission for all users
**C.** Adds execute permission for the owner only
**D.** Changes the owner of the file

### Solution — Question 2

**Correct Answer: C**

`u` refers to the user (owner), `+x` adds execute permission. So `chmod u+x` adds execute permission for the file owner only. Group and others are not affected.

**Why the other options are incorrect:**

- **A.** The `+` symbol adds permission; `-` would remove it (`u-x`).

### Key Takeaways

- Permissions are read(4), write(2), execute(1) for owner, group, others.
- `755` = rwxr-xr-x is standard for scripts; `644` = rw-r--r-- for regular files.
- `chmod` changes permissions; `chown` changes ownership.
- umask `022` means new files default to 644 and new directories to 755.
- Use symbolic notation (`u+x`) for quick changes, numeric (`755`) for exact settings.

## Process Scheduling and Memory Management (Mid-Level)

### Process Scheduling and Memory Management

**CPU Scheduling** determines which process runs on the CPU and for how long. The kernel's scheduler makes these decisions.

**Common scheduling algorithms:**
- **First-Come, First-Served (FCFS):** Processes run in arrival order. Simple but can cause the *convoy effect* (short jobs wait behind long ones).
- **Shortest Job First (SJF):** Runs the shortest job next. Optimal average wait time but requires knowing job lengths in advance.
- **Round Robin (RR):** Each process gets a fixed *time quantum* (e.g., 10ms). After the quantum expires, the process goes to the back of the queue. Fair and responsive.
- **Priority Scheduling:** Each process has a priority. Higher-priority processes run first. Risk of *starvation* (low-priority processes never run) — solved by *aging*.

**Linux uses the Completely Fair Scheduler (CFS)** — a weighted fair-queuing scheduler that gives each process a fair share of CPU time based on its nice value.

**Memory Management:**
- **Virtual Memory:** Each process sees its own address space, mapped to physical memory by the MMU (Memory Management Unit).
- **Paging:** Memory is divided into fixed-size *pages* (typically 4KB). The OS maps virtual pages to physical *frames*.
- **Page Fault:** Occurs when a process accesses a page not in physical memory. The OS loads it from disk (swap).
- **Swap:** Disk space used as overflow when physical RAM is full. Swapping is slow — excessive swapping is called *thrashing*.

````python
# --- Round Robin scheduling simulation ---
# Demonstrates how RR scheduling works with a time quantum

from collections import deque

def round_robin(processes: list[tuple[str, int]], quantum: int) -> list[str]:
    """
    Simulate Round Robin scheduling.
    processes: list of (name, burst_time) tuples
    quantum: time slice for each process
    Returns: execution order log
    """
    queue = deque([(name, burst) for name, burst in processes])
    log = []  # execution order
    time = 0

    while queue:
        name, remaining = queue.popleft()
        # Process runs for min(quantum, remaining) time
        run_time = min(quantum, remaining)
        time += run_time
        remaining -= run_time
        log.append(f'{name}(ran {run_time}ms, t={time}ms)')

        if remaining > 0:
            queue.append((name, remaining))  # back to queue

    return log

# Example: 3 processes with quantum = 4ms
processes = [('P1', 10), ('P2', 4), ('P3', 6)]
for entry in round_robin(processes, quantum=4):
    print(entry)
# P1(ran 4ms, t=4ms)
# P2(ran 4ms, t=8ms)    <- P2 finishes
# P3(ran 4ms, t=12ms)
# P1(ran 4ms, t=16ms)
# P3(ran 2ms, t=18ms)   <- P3 finishes
# P1(ran 2ms, t=20ms)   <- P1 finishes
````

## CI/CD Pipeline Concepts (Jenkins) (Mid-Level)

### CI/CD Pipeline Concepts — Jenkins (nip-Relevant)

**Continuous Integration (CI)** automatically builds and tests code on every commit. **Continuous Delivery/Deployment (CD)** automates releasing tested code to staging or production.

**Pipeline stages (typical flow):**
1. **Source** — code checkout from Git
2. **Build** — compile code, build Docker images
3. **Test** — run unit tests, integration tests, linting
4. **Deploy** — push to staging/production

**Jenkins** is a widely-used open-source CI/CD server:
- **Jenkinsfile** — defines the pipeline as code (Groovy DSL)
- **Declarative Pipeline** — structured syntax with `pipeline { }` block
- **Stages** — logical groupings of steps
- **Steps** — individual commands within a stage
- **Agents** — where the pipeline runs (any, docker, specific node)

**nip context:** nip uses Jenkins (and similar tools) for building drivers, running GPU test suites, and deploying firmware updates across server fleets.

````python
# --- Jenkinsfile example (declarative pipeline) ---
import textwrap

jenkinsfile = textwrap.dedent('''
    // Jenkinsfile — Declarative Pipeline
    pipeline {
        agent any

        environment {
            DOCKER_IMAGE = "myapp:${BUILD_NUMBER}"
        }

        stages {
            stage("Checkout") {
                steps {
                    git branch: "main", url: "https://github.com/org/repo.git"
                }
            }
            stage("Build") {
                steps {
                    sh "docker build -t ${DOCKER_IMAGE} ."
                }
            }
            stage("Test") {
                steps {
                    sh "docker run --rm ${DOCKER_IMAGE} pytest tests/"
                }
            }
            stage("Deploy") {
                when {
                    branch "main"
                }
                steps {
                    sh "docker push ${DOCKER_IMAGE}"
                    sh "kubectl set image deployment/myapp myapp=${DOCKER_IMAGE}"
                }
            }
        }
        post {
            failure {
                echo "Pipeline failed! Sending notification..."
            }
        }
    }
''')
print(jenkinsfile)
````

### Key Takeaways

## PXE Boot Overview (Mid-Level)

### PXE Boot Overview (nip-Relevant)

**PXE (Preboot Execution Environment)** allows a computer to boot from the network instead of a local disk. It's widely used for:
- Automated OS installation across server fleets
- Diskless workstations
- System recovery and diagnostics

**PXE boot flow:**
1. Client powers on and sends a DHCP Discover broadcast.
2. DHCP server responds with an IP address AND the PXE boot server address + boot filename.
3. Client downloads the boot loader (e.g., `pxelinux.0`) via **TFTP** (Trivial File Transfer Protocol).
4. Boot loader downloads the kernel and initrd from the TFTP server.
5. Kernel boots and may fetch additional files (kickstart/preseed) via HTTP/NFS for automated installation.

**Key components:**
- **DHCP server** — assigns IP and points to TFTP server
- **TFTP server** — serves the boot loader and kernel
- **HTTP/NFS server** — serves the OS installation files
- **Kickstart (RHEL) / Preseed (Debian)** — automated install config

**nip context:** PXE boot is used to provision bare-metal GPU servers at scale, deploy custom OS images with nip drivers, and manage server fleets in data centers.

````python
# --- PXE boot configuration examples ---
import textwrap

pxe_config = textwrap.dedent('''
    # --- DHCP server config (ISC DHCP) ---
    # /etc/dhcp/dhcpd.conf
    subnet 10.0.1.0 netmask 255.255.255.0 {
        range 10.0.1.100 10.0.1.200;
        option routers 10.0.1.1;
        option domain-name-servers 10.0.1.1;
        # PXE boot options
        next-server 10.0.1.10;          # TFTP server IP
        filename "pxelinux.0";           # Boot loader file
    }

    # --- PXE menu config ---
    # /tftpboot/pxelinux.cfg/default
    DEFAULT menu.c32
    PROMPT 0
    TIMEOUT 300
    MENU TITLE PXE Boot Menu

    LABEL ubuntu
        MENU LABEL Install Ubuntu 22.04
        KERNEL ubuntu/vmlinuz
        APPEND initrd=ubuntu/initrd.gz auto=true url=http://10.0.1.10/preseed.cfg

    LABEL centos
        MENU LABEL Install CentOS 9
        KERNEL centos/vmlinuz
        APPEND initrd=centos/initrd.img ks=http://10.0.1.10/kickstart.cfg

    # --- PXE boot flow summary ---
    # 1. Client: DHCP Discover (broadcast)
    # 2. DHCP Server: Offer IP + next-server=10.0.1.10, filename=pxelinux.0
    # 3. Client: TFTP GET pxelinux.0 from 10.0.1.10
    # 4. Client: TFTP GET pxelinux.cfg/default (menu config)
    # 5. Client: TFTP GET kernel + initrd
    # 6. Kernel boots, fetches kickstart/preseed via HTTP
    # 7. Automated OS installation begins
''')
print(pxe_config)
````

### Solution — Question 1

TFTP (Trivial File Transfer Protocol) is used for the initial boot loader transfer in PXE. It's a simple, lightweight protocol that works well in the pre-OS boot environment where a full TCP/IP stack isn't available yet.

- PXE is essential for provisioning bare-metal servers at scale in data centers.
