## Arrays and Strings (Beginner)

### Arrays and Strings

Arrays (Python `list`) and strings (`str`) are the most fundamental data structures. They store elements in contiguous memory, allowing O(1) random access by index.

**Key operations:**
- Indexing: `arr[i]` — O(1)
- Slicing: `arr[i:j]` — O(j - i)
- Append: `arr.append(x)` — amortized O(1)
- Insert at index: `arr.insert(i, x)` — O(n)
- Search (unsorted): `x in arr` — O(n)

**Common patterns:**
- Two-pointer technique
- Sliding window
- Frequency counting with dictionaries
- Reversing in-place

````python
# --- Arrays: basic operations ---
arr = [10, 20, 30, 40, 50]

# Indexing — O(1)
print(arr[2])          # 30

# Slicing — O(k) where k = slice length
print(arr[1:4])        # [20, 30, 40]

# Append — amortized O(1)
arr.append(60)
print(arr)             # [10, 20, 30, 40, 50, 60]

# Insert at index — O(n) due to shifting
arr.insert(0, 5)
print(arr)             # [5, 10, 20, 30, 40, 50, 60]

# Reverse in-place — O(n)
arr.reverse()
print(arr)             # [60, 50, 40, 30, 20, 10, 5]
````

````python
# --- Strings: common operations ---
s = 'hello world'

# Strings are immutable — slicing creates a new string
print(s[0:5])          # 'hello'
print(s[::-1])         # 'dlrow olleh'  (reverse)

# Frequency count using a dict
freq = {}
for ch in s:
    freq[ch] = freq.get(ch, 0) + 1
print(freq)            # {'h': 1, 'e': 1, 'l': 3, ...}

# Two-pointer: check if a string is a palindrome
def is_palindrome(text: str) -> bool:
    left, right = 0, len(text) - 1
    while left < right:
        if text[left] != text[right]:
            return False
        left += 1
        right -= 1
    return True

print(is_palindrome('racecar'))  # True
print(is_palindrome('hello'))    # False
````

## Linked Lists (Beginner)

### Linked Lists

A linked list is a linear data structure where each element (node) contains a value and a reference (pointer) to the next node. Unlike arrays, linked lists do not store elements contiguously in memory.

**Key operations:**
- Insert at head: O(1)
- Insert at tail (with tail pointer): O(1)
- Search: O(n)
- Delete by value: O(n)
- Traverse: O(n)

**Advantages over arrays:**
- O(1) insertion/deletion at the head
- Dynamic size — no need to pre-allocate

**Disadvantages:**
- No random access (must traverse from head)
- Extra memory for pointers

````python
# --- Singly Linked List implementation ---
class Node:
    """A single node in the linked list."""
    def __init__(self, data):
        self.data = data
        self.next = None  # pointer to next node

class SinglyLinkedList:
    """Singly linked list with insert, delete, search, traverse."""
    def __init__(self):
        self.head = None

    def insert_at_head(self, data):
        """Insert a new node at the beginning — O(1)."""
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_tail(self, data):
        """Insert a new node at the end — O(n)."""
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def search(self, target):
        """Search for a value — O(n)."""
        current = self.head
        while current:
            if current.data == target:
                return True
            current = current.next
        return False

    def delete(self, target):
        """Delete first occurrence of target — O(n)."""
        if not self.head:
            return
        if self.head.data == target:
            self.head = self.head.next
            return
        current = self.head
        while current.next:
            if current.next.data == target:
                current.next = current.next.next
                return
            current = current.next

    def traverse(self):
        """Print all elements — O(n)."""
        elements = []
        current = self.head
        while current:
            elements.append(current.data)
            current = current.next
        return elements

# Usage
ll = SinglyLinkedList()
ll.insert_at_head(3)
ll.insert_at_head(2)
ll.insert_at_head(1)
ll.insert_at_tail(4)
print(ll.traverse())    # [1, 2, 3, 4]
print(ll.search(3))     # True
ll.delete(2)
print(ll.traverse())    # [1, 3, 4]
````

## Stacks and Queues (Beginner)

### Stacks and Queues

**Stack (LIFO — Last In, First Out):** Think of a stack of plates. The last plate placed on top is the first one removed.

**Queue (FIFO — First In, First Out):** Think of a line at a store. The first person in line is the first one served.

**Stack operations:**
- `push(x)` — add to top: O(1)
- `pop()` — remove from top: O(1)
- `peek()` — view top element: O(1)

**Queue operations:**
- `enqueue(x)` — add to rear: O(1)
- `dequeue()` — remove from front: O(1)
- `peek()` — view front element: O(1)

In Python, use a `list` for stacks and `collections.deque` for queues (deque provides O(1) popleft, whereas list.pop(0) is O(n)).

````python
# --- Stack using a Python list ---
stack = []

# Push — O(1) amortized
stack.append(10)
stack.append(20)
stack.append(30)
print(stack)           # [10, 20, 30]

# Peek — O(1)
print(stack[-1])       # 30  (top of stack)

# Pop — O(1) amortized
top = stack.pop()
print(top)             # 30
print(stack)           # [10, 20]

# Check if empty
print(len(stack) == 0) # False
````

````python
# --- Queue using collections.deque ---
from collections import deque

queue = deque()

# Enqueue — O(1)
queue.append(10)
queue.append(20)
queue.append(30)
print(queue)           # deque([10, 20, 30])

# Peek front — O(1)
print(queue[0])        # 10

# Dequeue — O(1)
front = queue.popleft()
print(front)           # 10
print(queue)           # deque([20, 30])

# Common use: BFS traversal uses a queue
# Common use: DFS / backtracking uses a stack
````

- Stacks are used in DFS, expression evaluation, and undo systems.
- Queues are used in BFS, task scheduling, and buffering.

## Hash Tables (Beginner)

### Hash Tables

A hash table (Python `dict`) maps keys to values using a hash function. The hash function converts a key into an index in an internal array, enabling average-case O(1) lookups.

**Key operations (average case):**
- Insert: O(1)
- Lookup: O(1)
- Delete: O(1)

**Worst case:** O(n) when many keys hash to the same bucket (hash collision).

**Collision handling strategies:**
- **Chaining:** Each bucket stores a linked list of entries.
- **Open addressing:** On collision, probe the next available slot (linear probing, quadratic probing, double hashing).

**Python dict internals:**
- Uses open addressing with a compact hash table.
- Keys must be hashable (immutable types: `int`, `str`, `tuple`).
- Maintains insertion order (Python 3.7+).

````python
# --- Hash Table (dict) operations ---
# Creating a dict
phonebook = {'Alice': '555-0100', 'Bob': '555-0200'}

# Insert / Update — O(1) average
phonebook['Charlie'] = '555-0300'
print(phonebook)
# {'Alice': '555-0100', 'Bob': '555-0200', 'Charlie': '555-0300'}

# Lookup — O(1) average
print(phonebook['Alice'])   # '555-0100'
print(phonebook.get('Dave', 'Not found'))  # 'Not found'

# Delete — O(1) average
del phonebook['Bob']
print(phonebook)  # {'Alice': '555-0100', 'Charlie': '555-0300'}

# Check membership — O(1) average
print('Alice' in phonebook)  # True
````

````python
# --- Common pattern: frequency counting ---
def most_frequent(nums: list[int]) -> int:
    """Return the most frequent element. O(n) time, O(n) space."""
    freq = {}  # hash table for counting
    for num in nums:
        freq[num] = freq.get(num, 0) + 1
    # Find key with max value
    return max(freq, key=freq.get)

print(most_frequent([1, 3, 2, 3, 3, 1]))  # 3

# --- Two-sum using a hash table ---
def two_sum(nums: list[int], target: int) -> list[int]:
    """Return indices of two numbers that add up to target. O(n)."""
    seen = {}  # value -> index
    for i, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            return [seen[complement], i]
        seen[num] = i
    return []

print(two_sum([2, 7, 11, 15], 9))  # [0, 1]
````

## Basic Sorting (Beginner)

### Basic Sorting Algorithms

Sorting is a fundamental operation. Understanding basic O(n²) sorts helps build intuition before tackling efficient O(n log n) algorithms.

| Algorithm | Best | Average | Worst | Space | Stable? |
|-----------|------|---------|-------|-------|---------|
| Bubble Sort | O(n) | O(n²) | O(n²) | O(1) | Yes |
| Selection Sort | O(n²) | O(n²) | O(n²) | O(1) | No |
| Insertion Sort | O(n) | O(n²) | O(n²) | O(1) | Yes |

**Bubble Sort:** Repeatedly swap adjacent elements if they're in the wrong order. Best case O(n) when already sorted (with early-exit optimization).

**Selection Sort:** Find the minimum element in the unsorted portion and swap it to the front. Always O(n²) comparisons.

**Insertion Sort:** Build the sorted array one element at a time by inserting each element into its correct position. Efficient for nearly-sorted data.

````python
# --- Bubble Sort ---
# Time: O(n²) average/worst, O(n) best (already sorted)
# Space: O(1) — in-place
def bubble_sort(arr: list) -> list:
    """Sort array using bubble sort with early-exit optimization."""
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]  # swap
                swapped = True
        if not swapped:
            break  # already sorted — early exit
    return arr

print(bubble_sort([64, 34, 25, 12, 22, 11, 90]))
# [11, 12, 22, 25, 34, 64, 90]
````

````python
# --- Selection Sort ---
# Time: O(n²) always
# Space: O(1) — in-place
def selection_sort(arr: list) -> list:
    """Sort array by repeatedly selecting the minimum."""
    n = len(arr)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]  # swap minimum to front
    return arr

print(selection_sort([64, 25, 12, 22, 11]))
# [11, 12, 22, 25, 64]
````

````python
# --- Insertion Sort ---
# Time: O(n²) average/worst, O(n) best (nearly sorted)
# Space: O(1) — in-place
def insertion_sort(arr: list) -> list:
    """Sort array by inserting each element into its correct position."""
    for i in range(1, len(arr)):
        key = arr[i]  # element to insert
        j = i - 1
        # Shift elements greater than key to the right
        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = key  # place key in correct position
    return arr

print(insertion_sort([12, 11, 13, 5, 6]))
# [5, 6, 11, 12, 13]
````

## Binary Trees and BSTs (Mid-Level)

### Binary Trees and Binary Search Trees (BSTs)

A **binary tree** is a tree where each node has at most two children (left and right). A **Binary Search Tree (BST)** adds an ordering property: for every node, all values in the left subtree are smaller and all values in the right subtree are larger.

**BST operations (average case — balanced tree):**
- Search: O(log n)
- Insert: O(log n)
- Delete: O(log n)

**Worst case (degenerate/skewed tree):** O(n) for all operations.

**Tree traversals:**
- **In-order (Left, Root, Right):** Produces sorted output for BSTs.
- **Pre-order (Root, Left, Right):** Useful for copying/serializing trees.
- **Post-order (Left, Right, Root):** Useful for deleting trees.

````python
# --- Binary Search Tree implementation ---
class TreeNode:
    """A node in the binary search tree."""
    def __init__(self, val):
        self.val = val
        self.left = None
        self.right = None

class BST:
    """Binary Search Tree with insert, search, and traversals."""
    def __init__(self):
        self.root = None

    def insert(self, val):
        """Insert a value into the BST — O(log n) average."""
        self.root = self._insert(self.root, val)

    def _insert(self, node, val):
        if node is None:
            return TreeNode(val)
        if val < node.val:
            node.left = self._insert(node.left, val)
        elif val > node.val:
            node.right = self._insert(node.right, val)
        return node  # duplicates ignored

    def search(self, val) -> bool:
        """Search for a value — O(log n) average."""
        return self._search(self.root, val)

    def _search(self, node, val) -> bool:
        if node is None:
            return False
        if val == node.val:
            return True
        elif val < node.val:
            return self._search(node.left, val)
        else:
            return self._search(node.right, val)

    def inorder(self) -> list:
        """In-order traversal — returns sorted values."""
        result = []
        self._inorder(self.root, result)
        return result

    def _inorder(self, node, result):
        if node:
            self._inorder(node.left, result)
            result.append(node.val)
            self._inorder(node.right, result)

    def preorder(self) -> list:
        """Pre-order traversal (Root, Left, Right)."""
        result = []
        self._preorder(self.root, result)
        return result

    def _preorder(self, node, result):
        if node:
            result.append(node.val)
            self._preorder(node.left, result)
            self._preorder(node.right, result)

# Usage
bst = BST()
for val in [5, 3, 7, 1, 4, 6, 8]:
    bst.insert(val)

print(bst.inorder())    # [1, 3, 4, 5, 6, 7, 8]  (sorted!)
print(bst.preorder())   # [5, 3, 1, 4, 7, 6, 8]
print(bst.search(4))    # True
print(bst.search(9))    # False
````

## Graphs and BFS/DFS (Mid-Level)

### Graphs and BFS/DFS

A **graph** consists of vertices (nodes) and edges connecting them. Graphs can be directed or undirected, weighted or unweighted.

**Representation — Adjacency List:**
The most common representation uses a dictionary where each key is a vertex and the value is a list of its neighbors. Space: O(V + E).

**Breadth-First Search (BFS):**
- Explores level by level using a queue.
- Time: O(V + E), Space: O(V)
- Use cases: shortest path (unweighted), level-order traversal.

**Depth-First Search (DFS):**
- Explores as deep as possible before backtracking, using a stack (or recursion).
- Time: O(V + E), Space: O(V)
- Use cases: cycle detection, topological sort, connected components.

````python
# --- Graph representation and BFS ---
from collections import deque

# Adjacency list representation
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E'],
}

def bfs(graph: dict, start: str) -> list[str]:
    """Breadth-First Search — O(V + E) time, O(V) space."""
    visited = set()
    queue = deque([start])
    visited.add(start)
    order = []

    while queue:
        vertex = queue.popleft()  # dequeue front
        order.append(vertex)
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)  # enqueue
    return order

print(bfs(graph, 'A'))  # ['A', 'B', 'C', 'D', 'E', 'F']
````

````python
# --- DFS (iterative with stack) ---
def dfs_iterative(graph: dict, start: str) -> list[str]:
    """Depth-First Search using a stack — O(V + E) time."""
    visited = set()
    stack = [start]
    order = []

    while stack:
        vertex = stack.pop()  # pop from top (LIFO)
        if vertex not in visited:
            visited.add(vertex)
            order.append(vertex)
            # Add neighbors in reverse for consistent ordering
            for neighbor in reversed(graph[vertex]):
                if neighbor not in visited:
                    stack.append(neighbor)
    return order

# --- DFS (recursive) ---
def dfs_recursive(graph: dict, start: str, visited: set = None) -> list[str]:
    """Depth-First Search using recursion — O(V + E) time."""
    if visited is None:
        visited = set()
    visited.add(start)
    order = [start]
    for neighbor in graph[start]:
        if neighbor not in visited:
            order.extend(dfs_recursive(graph, neighbor, visited))
    return order

graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E'],
}
print(dfs_iterative(graph, 'A'))  # ['A', 'B', 'D', 'E', 'F', 'C']
print(dfs_recursive(graph, 'A'))  # ['A', 'B', 'D', 'E', 'F', 'C']
````

### Solution — Question 1

**Correct Answer: B**

## Advanced Sorting (Mid-Level)

### Advanced Sorting — Merge Sort and Quick Sort

These divide-and-conquer algorithms achieve O(n log n) average-case performance, making them practical for large datasets.

| Algorithm | Best | Average | Worst | Space | Stable? |
|-----------|------|---------|-------|-------|---------|
| Merge Sort | O(n log n) | O(n log n) | O(n log n) | O(n) | Yes |
| Quick Sort | O(n log n) | O(n log n) | O(n²) | O(log n) | No |

**Merge Sort:** Divide the array in half, recursively sort each half, then merge the two sorted halves. Guaranteed O(n log n) but uses O(n) extra space.

**Quick Sort:** Pick a pivot, partition elements into those less than and greater than the pivot, then recursively sort each partition. O(n log n) average but O(n²) worst case (poor pivot choice). In-place with O(log n) stack space.

**Python's built-in `sorted()` and `list.sort()`** use Timsort, a hybrid of Merge Sort and Insertion Sort — O(n log n) worst case, stable, and highly optimized.

````python
# --- Merge Sort ---
# Time: O(n log n) always
# Space: O(n) — needs auxiliary arrays
def merge_sort(arr: list) -> list:
    """Sort array using merge sort (divide and conquer)."""
    if len(arr) <= 1:
        return arr

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])    # sort left half
    right = merge_sort(arr[mid:])   # sort right half
    return merge(left, right)

def merge(left: list, right: list) -> list:
    """Merge two sorted arrays into one sorted array."""
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:  # <= for stability
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])   # remaining elements
    result.extend(right[j:])
    return result

print(merge_sort([38, 27, 43, 3, 9, 82, 10]))
# [3, 9, 10, 27, 38, 43, 82]
````

````python
# --- Quick Sort ---
# Time: O(n log n) average, O(n²) worst case
# Space: O(log n) stack space
def quick_sort(arr: list) -> list:
    """Sort array using quick sort with last-element pivot."""
    if len(arr) <= 1:
        return arr

    pivot = arr[-1]  # choose last element as pivot
    left = [x for x in arr[:-1] if x <= pivot]
    right = [x for x in arr[:-1] if x > pivot]
    return quick_sort(left) + [pivot] + quick_sort(right)

print(quick_sort([10, 7, 8, 9, 1, 5]))
# [1, 5, 7, 8, 9, 10]
````

## Searching Algorithms (Mid-Level)

### Searching Algorithms — Binary Search

**Binary Search** works on sorted arrays by repeatedly dividing the search space in half. It's one of the most important algorithms to master for interviews.

**Time complexity:** O(log n)
**Space complexity:** O(1) iterative, O(log n) recursive

**Variants:**
- **Standard:** Find exact target.
- **Lower bound (bisect_left):** Find the first position where target could be inserted to keep the array sorted.
- **Upper bound (bisect_right):** Find the position after the last occurrence of target.

**Key insight:** Binary search can be applied to any monotonic function, not just sorted arrays (e.g., 'find the minimum speed to finish within time T').

````python
# --- Standard Binary Search ---
# Time: O(log n), Space: O(1)
def binary_search(arr: list, target: int) -> int:
    """Return index of target, or -1 if not found."""
    left, right = 0, len(arr) - 1

    while left <= right:
        mid = (left + right) // 2  # avoid overflow in other languages
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1   # search right half
        else:
            right = mid - 1  # search left half
    return -1  # not found

arr = [1, 3, 5, 7, 9, 11, 13]
print(binary_search(arr, 7))   # 3
print(binary_search(arr, 4))   # -1
````

````python
# --- Lower Bound and Upper Bound ---
def lower_bound(arr: list, target: int) -> int:
    """Find first index where arr[i] >= target. O(log n)."""
    left, right = 0, len(arr)
    while left < right:
        mid = (left + right) // 2
        if arr[mid] < target:
            left = mid + 1
        else:
            right = mid
    return left

def upper_bound(arr: list, target: int) -> int:
    """Find first index where arr[i] > target. O(log n)."""
    left, right = 0, len(arr)
    while left < right:
        mid = (left + right) // 2
        if arr[mid] <= target:
            left = mid + 1
        else:
            right = mid
    return left

arr = [1, 3, 3, 3, 5, 7]
print(lower_bound(arr, 3))  # 1  (first index of 3)
print(upper_bound(arr, 3))  # 4  (first index after all 3s)
# Count of 3s = upper_bound - lower_bound = 4 - 1 = 3
````

## Big-O Complexity Analysis (Mid-Level)

### Big-O Complexity Analysis

Big-O notation describes the upper bound of an algorithm's growth rate as input size increases. It tells you how an algorithm **scales**, not its exact runtime.

**Common complexities (fastest to slowest):**

| Big-O | Name | Example |
|-------|------|---------|
| O(1) | Constant | Array index access, hash table lookup |
| O(log n) | Logarithmic | Binary search |
| O(n) | Linear | Single loop through array |
| O(n log n) | Linearithmic | Merge sort, efficient sorting |
| O(n²) | Quadratic | Nested loops, bubble sort |
| O(2ⁿ) | Exponential | Recursive Fibonacci (naive) |
| O(n!) | Factorial | Generating all permutations |

**Rules for analyzing code:**
1. **Drop constants:** O(2n) → O(n)
2. **Drop lower-order terms:** O(n² + n) → O(n²)
3. **Nested loops multiply:** Two nested loops over n → O(n²)
4. **Sequential code adds:** Loop O(n) then loop O(m) → O(n + m)
5. **Recursive calls:** Analyze the recurrence relation (e.g., T(n) = 2T(n/2) + O(n) → O(n log n) for merge sort)

````python
# --- Analyzing time complexity ---

# O(1) — Constant time
def get_first(arr):
    return arr[0]  # single operation, independent of n

# O(n) — Linear time
def find_max(arr):
    max_val = arr[0]
    for x in arr:        # one loop through n elements
        if x > max_val:
            max_val = x
    return max_val

# O(n²) — Quadratic time
def has_duplicate_pair(arr):
    for i in range(len(arr)):       # outer loop: n
        for j in range(i + 1, len(arr)):  # inner loop: ~n
            if arr[i] == arr[j]:    # n * n = n²
                return True
    return False

# O(log n) — Logarithmic time
def count_halvings(n):
    count = 0
    while n > 1:
        n //= 2   # halving each iteration → log₂(n) iterations
        count += 1
    return count

print(count_halvings(16))  # 4  (16 → 8 → 4 → 2 → 1)
````

````python
# --- Space complexity analysis ---

# O(1) space — in-place
def reverse_in_place(arr):
    left, right = 0, len(arr) - 1
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]
        left += 1
        right -= 1
    # Only uses a constant number of variables

# O(n) space — creates new data structure
def get_squares(arr):
    return [x ** 2 for x in arr]  # new list of size n

# O(n) space — recursive call stack
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)  # n frames on the call stack
````
